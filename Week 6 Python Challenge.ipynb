```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np 

figure = plt.figure() # create an object called figure and add a third axis to it
axis = figure.add_subplot(111, projection = '3d')

u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j] # mgrid creates a mesh-grid, which is an array that contains other arrays
# first argument interpretation: start from 0 and go to 2pi by 20j which is a complex number
# if step number is a complex number, the end value is included

# What we've done is created angles u and v that we are going to use to make our sphere

# Create a wireframe sphere
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
axis.plot_wireframe(x, y, z, color="r")
```
