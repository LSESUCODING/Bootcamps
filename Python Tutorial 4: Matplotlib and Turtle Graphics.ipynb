```python
import turtle

# Create a title for your canvas window
turtle.title("Turtle Animation")

# Set the screen size in pixels and the starting point of the turtle
turtle.setup(425, 425, 0, 0)

# Moving the turtle forward: argument is number of pixels to move by
turtle.forward(50)

# Turn your turtle: argument is number of degrees to turn by
turtle.right(180)
turtle.left(90)

turtle.forward(50)
turtle.right(45)

turtle.goto(0,0)
```

```python
# Stop and start drawing
turtle.penup() # stop drawing
turtle.goto(50,-100)
turtle.pendown() # start drawing again

turtle.forward(50)
```

```python
# Pen color
turtle.pencolor("#FF0000")
    
# Fill shapes
turtle.begin_fill() # tells Python to start filling in your shape before you start drawing

turtle.right(90)
turtle.forward(50)
turtle.goto(50,-100)

turtle.fillcolor("#FF0000") # choose the color
turtle.end_fill() # stop filling in your shape
```

```python
# Close your turtle window once you click on it
turtle.exitonclick()
```
